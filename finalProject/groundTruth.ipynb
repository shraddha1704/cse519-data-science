```python
def add_feat(gd, col):
    col13 = col + '_2013'
    col14 = col + '_2014'
    col15 = col + '_2015'
    col16 = col + '_2016'
    col17 = col + '_2017'

    conditions  = [ gd['year'] == 2013, gd['year'] == 2014, gd['year'] == 2015, gd['year'] == 2016, gd['year'] == 2017 ]
    choices     = [ gd[col13], gd[col14], gd[col15], gd[col16], gd[col17] ]

    gd[col] = np.select(conditions, choices, default=np.nan)

    gd.drop([col13,col14,col15,col16,col17], inplace=True, axis=1)
    return gd
```

```python
gd = gd[gd['year'].isin([2013,2014,2015,2016,2017])]



gd = add_feat(gd, 'Civilian_labor_force')
gd = add_feat(gd, 'Employed')
gd = add_feat(gd, 'Unemployed')
gd = add_feat(gd, 'Unemployment_rate')
gd = add_feat(gd, 'POP_ESTIMATE')
gd = add_feat(gd, 'N_POP_CHG')
gd = add_feat(gd, 'Births')
gd = add_feat(gd, 'Deaths')
gd = add_feat(gd, 'NATURAL_INC')
gd = add_feat(gd, 'INTERNATIONAL_MIG')
gd = add_feat(gd, 'DOMESTIC_MIG')
gd = add_feat(gd, 'NET_MIG')
gd = add_feat(gd, 'RESIDUAL')
gd = add_feat(gd, 'GQ_ESTIMATES')
gd = add_feat(gd, 'R_birth')
gd = add_feat(gd, 'R_death')
gd = add_feat(gd, 'R_NATURAL_INC')
gd = add_feat(gd, 'R_INTERNATIONAL_MIG')
gd = add_feat(gd, 'R_DOMESTIC_MIG')
gd = add_feat(gd, 'R_NET_MIG')
```
